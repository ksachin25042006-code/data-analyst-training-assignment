```python
import pandas as pd
import numpy as np

print("🌾 Initializing Precision Crop Yield Optimization System...")

# Set random seed for consistent data generation
np.random.seed(42)
num_records = 500

# 1. Generate Soil & Climate Data Metrics
data = {
    'region_id': [f"Region_{i:03d}" for i in range(num_records)],
    'nitrogen_content': np.random.uniform(20, 140, num_records),     # N levels in soil
    'phosphorus_content': np.random.uniform(10, 100, num_records),   # P levels in soil
    'potassium_content': np.random.uniform(10, 200, num_records),    # K levels in soil
    'temperature_celsius': np.random.uniform(15, 38, num_records),   # Regional Temperature
    'humidity_percentage': np.random.uniform(30, 90, num_records),   # Regional Humidity
    'rainfall_mm': np.random.uniform(50, 300, num_records),          # Total Rainfall
    'soil_ph': np.random.uniform(5.0, 8.5, num_records)              # Soil Acidity
}

df_agri = pd.DataFrame(data)

# Calculate synthetic Crop Yield based on physical parameters (Regression Target)
# Yield increases with optimal rainfall, balanced pH, and high nitrogen nutrients
df_agri['crop_yield_mt'] = (
    (df_agri['nitrogen_content'] * 0.4) +
    (df_agri['rainfall_mm'] * 0.3) -
    (abs(df_agri['soil_ph'] - 6.5) * 15) +
    np.random.normal(20, 5, num_records)
).clip(lower=2) # Clip to ensure no negative yields

print(f"✅ Soil and Climate baseline established. Total regions mapped: {df_agri.shape[0]}")

# 2. Generate Farmer Forum / Market News Data for NLP Analysis
news_samples = [
    "Severe drought expected this month, crop yield risk is extremely high and fertilizer costs are soaring.",
    "Excellent rainfall this week! Soil moisture looks perfect for a bumper harvest.",
    "Pest outbreak reported in neighboring sectors, farmers complain about lack of government support.",
    "Market prices for wheat are collapsing, heavy financial loss expected for rural communities.",
    "Normal weather conditions, stable market demand and smooth supply chain movements.",
    "High heatwaves are drying up the soil, crop health is deteriorating rapidly."
]

news_data = []
for i in range(num_records):
    region = f"Region_{i:03d}"
    text = np.random.choice(news_samples)
    news_data.append({
        'region_id': region,
        'forum_post': text,
        'market_views': np.random.randint(100, 5000),
        'post_likes': np.random.randint(5, 250)
    })

df_news = pd.DataFrame(news_data)
print(f"✅ Text Analytics text corpus ready. Total social tokens: {df_news.shape[0]}")
```

```python
print("⚙️ Processing Dependent Variable Rules...")

# Apply multi-condition grouping to establish structural classification brackets
conditions = [
    (df_agri['crop_yield_mt'] < 45),
    (df_agri['crop_yield_mt'] >= 45) & (df_agri['crop_yield_mt'] <= 75),
    (df_agri['crop_yield_mt'] > 75)
]
choices = ['High Risk', 'Medium Risk', 'Low Risk']

df_agri['Yield_Risk_Tier'] = np.select(conditions, choices, default='Medium Risk')

print("✅ Target metrics computed successfully!")
print(df_agri[['region_id', 'crop_yield_mt', 'Yield_Risk_Tier']].head(4))
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
from statsmodels.stats.outliers_influence import variance_inflation_factor

print("📊 Starting Feature Engineering & Statistical Diagnostics...")

# Generate a synthetic Sentiment_Score from df_news
df_news['Sentiment_Score'] = df_news['post_likes'] / (df_news['market_views'] + 1) # Add 1 to avoid division by zero
# Clip the sentiment score to be in a readable range (e.g., 0 to 1)
df_news['Sentiment_Score'] = np.clip(df_news['Sentiment_Score'], 0, 1)

# Merge df_agri and df_news to create df_final
df_final = pd.merge(df_agri, df_news[['region_id', 'Sentiment_Score']], on='region_id', how='left')

# 1. Select numeric features for engineering analysis
numeric_cols = [
    'nitrogen_content', 'phosphorus_content', 'potassium_content',
    'temperature_celsius', 'humidity_percentage', 'rainfall_mm',
    'soil_ph', 'Sentiment_Score'
]
X_diag = df_final[numeric_cols].copy()

# 2. Check for Multicollinearity using Variance Inflation Factor (VIF)
vif_data = pd.DataFrame()
vif_data["Feature"] = X_diag.columns
vif_data["VIF_Value"] = [variance_inflation_factor(X_diag.values, i) for i in range(len(X_diag.columns))]

print("\n⚙️ Variance Inflation Factor (VIF) Matrix:")
print(vif_data.round(2))

# 3. Form a simple correlation table to show the interviewer
corr_matrix = X_diag.corr()['Sentiment_Score'].to_frame().reset_index()
print("\n🔗 Correlation to Public Sentiment Vector:")
print(corr_matrix.round(3))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
from sklearn.metrics import r2_score, mean_absolute_error, accuracy_score, classification_report

print("🤖 Initializing Machine Learning Engine...")

# 1. Feature Selection and Preprocessing
features = ['nitrogen_content', 'phosphorus_content', 'potassium_content',
            'temperature_celsius', 'humidity_percentage', 'rainfall_mm', 'soil_ph', 'Sentiment_Score']

X = df_final[features]
y_reg = df_final['crop_yield_mt']      # Regression Target
y_clf = df_final['Yield_Risk_Tier']     # Classification Target

# Scale features to handle high VIF multicollinearity
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Encode categorical targets
le = LabelEncoder()
y_clf_encoded = le.fit_transform(y_clf)

# 2. Split Data into Train/Test Sets
X_train_r, X_test_r, y_train_r, y_test_r = train_test_split(X_scaled, y_reg, test_size=0.2, random_state=42)
X_train_c, X_test_c, y_train_c, y_test_c = train_test_split(X_scaled, y_clf_encoded, test_size=0.2, random_state=42)

# 3. Train Regression Model (Random Forest Regressor)
reg_model = RandomForestRegressor(n_estimators=100, random_state=42)
reg_model.fit(X_train_r, y_train_r)
reg_preds = reg_model.predict(X_test_r)

print("\n📈 [REGRESSION MODEL RESULTS] (Target: Crop Yield)")
print(f"Random Forest Regressor R² Score: {r2_score(y_test_r, reg_preds):.3f}")
print(f"Mean Absolute Error (MAE): {mean_absolute_error(y_test_r, reg_preds):.2f} Metric Tons")

# 4. Train Classification Model (Random Forest Classifier)
clf_model = RandomForestClassifier(n_estimators=100, random_state=42)
clf_model.fit(X_train_c, y_train_c)
clf_preds = clf_model.predict(X_test_c)

print("\n🎯 [CLASSIFICATION MODEL RESULTS] (Target: Yield Risk Level)")
print(f"Random Forest Classifier Accuracy: {accuracy_score(y_test_c, clf_preds) * 100:.2f}%")
print("\nClassification Matrix Highlights:")
print(classification_report(y_test_c, clf_preds, target_names=le.classes_))
```

```python
%%writefile app.py
import streamlit as st
import pandas as pd
import numpy as np
import plotly.express as px

# 1. Setup App Headers
st.set_page_config(page_title="Precision Agri Intelligence", layout="wide")
st.title("🌾 Precision Crop Yield Optimization & Market Intelligence Dashboard")
st.markdown("---")

# 2. Sidebar Configuration Controls
st.sidebar.header("🕹️ Regional Simulation Parameters")
nitrogen = st.sidebar.slider("Soil Nitrogen Level (N)", 20, 140, 80)
rainfall = st.sidebar.slider("Total Seasonal Rainfall (mm)", 50, 300, 175)
soil_ph = st.sidebar.slider("Soil Acidity (pH Range)", 4.5, 8.5, 6.5)

# 3. Formulate Interactive Metrics Layout
col1, col2, col3 = st.columns(3)

# Simulated backend analytic calculation rule
predicted_yield = (nitrogen * 0.4) + (rainfall * 0.3) - (abs(soil_ph - 6.5) * 15) + 50
predicted_yield = max(2.0, predicted_yield)

if predicted_yield < 45:
    risk_level = "🚨 High Risk"
elif predicted_yield <= 75:
    risk_level = "⚠️ Medium Risk"
else:
    risk_level = "✅ Low Risk Zone"

with col1:
    st.metric(label="📊 Simulated Forecasted Crop Yield", value=f"{predicted_yield:.2f} MT / Acre")
with col2:
    st.metric(label="🛡️ Regional Production Risk Category", value=risk_level)
with col3:
    st.metric(label="💡 Public Sentiment Vector", value="Optimistic (+0.42)")

# 4. Interactive Simulation Visual Chart Elements
st.markdown("### 📈 Analytical Factors Trend Visualization")

# Create simple historical trend line to simulate data exploration dashboard
chart_data = pd.DataFrame({
    'Rainfall (mm)': np.linspace(50, 300, 50),
    'Expected Production Volume': np.linspace(50, 300, 50) * 0.35 + (nitrogen * 0.1)
})
fig = px.line(chart_data, x='Rainfall (mm)', y='Expected Production Volume', title='Production Sensitivity Scale vs Rainfall Patterns')
st.plotly_chart(fig, use_container_width=True)

print("📝 File 'app.py' successfully drafted inside your workspace root folder!")
```
